```python
import numpy as np
N = int(input())
row = list(input().split())
count = -1


def count_in_row(row:list):
    count_fives = 0
    if '2'in row or '3' in row:
        pass
    else:
        for number in row:
            if number == '5':
                count_fives +=1
    return count_fives

if N<7:
    count = count_in_row(row)
else:
    for i in range(7,N+1):
        print(i,row[i-7:i])
        answer = count_in_row(row[i-7:i])
        print(answer)
        if answer > count:
            count = answer
print(count)


    

```
