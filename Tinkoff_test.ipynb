Задача 1. Поиск наилучшего непрерывного отрезка с точки зрения школьных оценок

```python
N = int(input())
row = list(input().split())
count = -1


def count_in_row(row:list):
    count_fives = 0
    if '2'in row or '3' in row:
        count_fives = count
    else:
        for number in row:
            if number == '5':
                count_fives +=1
    return count_fives

if N<7:
    pass
else:
    for i in range(7,N+1):
        answer = count_in_row(row[i-7:i])
        if answer > count:
            count = answer
print(count)

    

```

Задача 2. Поворот матрицы на 90 градусов

```python
n, m = map(int,input().split())
matrix = []
for i in range(n):
    matrix.append(list(map(int,input().split())))

for j in range(m):
    for i in reversed(range(n)):
        print(matrix[i][j], end=" ")
    print("\n", end="")
```

Задача 3. Представление директорий


```python
N = int(input())
folders = []
for i in range(N):
    folder = list(input().split(sep = '/'))
    folders.append((folder))
folders = sorted(folders)
for i in range(len(folders)):
    print(("  "*(len(folders[i])-1)) + str(folders[i][len(folders[i])-1]))
```

Задача 5. Поиск наилучшего прохода через лес, для сбора максимального количества грибов.

```python
N = int(input())
answer = 0
forest = []
for i in range(N):
    forest.append(list('W'+str(input())+'W'))


def go_next(step: int,position: str,shrooms: int, cell:int ):
    global answer
    if position == '.':
        if step != N-1:
            for i in range(-1,2):
                go_next(step = step+1,position = forest[step+1][cell+i],shrooms = shrooms,cell = cell+i)
        
    elif position == 'C':
        shrooms +=1
        if step != N-1:
            for i in range(-1,2):
                go_next(step = step+1,position = forest[step+1][cell+i],shrooms = shrooms,cell = cell+i)
    if answer < shrooms:
        answer = shrooms



for j in range(1,4):
    go_next(0,forest[0][j],0,j)
print(answer)
```

задача 4. Поворот матрицы с помощью попарных перестановок.

```python

def rotate_matrix(matrix, direction):
    n = len(matrix)
    operations = []

    def swap(x1, y1, x2, y2):
        operations.append((x1, y1, x2, y2))
        matrix[x1][y1], matrix[x2][y2] = matrix[x2][y2], matrix[x1][y1]

    if direction == 'R':
        for i in range(n):
            for j in range(i + 1, n):
                swap(i, j, j, i)

        for i in range(n):
            for j in range(n // 2):
                swap(i, j, i, n - j - 1)

    elif direction == 'L':
        for i in range(n // 2):
            for j in range(n):
                swap(i, j, n - i - 1, j)

        for i in range(n):
            for j in range(i + 1, n):
                swap(i, j, j, i)

    return operations

n, direction = input().split()
n = int(n)
matrix = [list(map(int, input().split())) for _ in range(n)]

operations = rotate_matrix(matrix, direction)
print(len(operations))
for op in operations:
    print(*op)
```
